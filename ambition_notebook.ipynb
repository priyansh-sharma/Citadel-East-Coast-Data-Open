```python
import matplotlib.pyplot as plt
import pandas as pd

seds = pd.read_csv("seds.csv")
states_w_policies = pd.read_csv("states_with_policies.csv")




def re_prop(state):
    state_RETCB = seds.loc[(seds.state_name == state) & (seds.msn == 'RETCB')]
    state_TETCB = seds.loc[(seds.state_name == state) & (seds.msn == 'TETCB')]

    # Create a plot with a caption, X and Y legends, etc
    x_label = 'Year'
    y_label = 'RETCB/TETCB'
    plt.title('RE/TE')
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    plt.scatter(state_RETCB.year, (state_RETCB.value.values)/(state_TETCB.value.values), color='red', label='Points')
    plt.legend()

    plt.show()
    
    out = pd.DataFrame(state_RETCB.year)
    out['RETCB'] = state_RETCB.value.values
    out['TETCB'] = state_TETCB.value.values
    return out

    
states_w_policies.drop(states_w_policies.index[36], inplace=True)


important_cols = states_w_policies[["Unnamed: 0.1", "first_target_pct", "year_enacted", "first_target_year", "nearest_year"]]
important_cols
```

```python
important_cols["result"] = 0.0

for i in important_cols.index:
    state = important_cols.loc[i, "Unnamed: 0.1"]
    year = important_cols.loc[i, "nearest_year"]

    print(state)
    state_array = re_prop(state)
    row = state_array[state_array.year == year]

    ratio = row["RETCB"]/row["TETCB"]
    # print(ratio.value)

    #display(important_cols)
    important_cols.at[i,'result'] = ratio
    display(important_cols)
```

```python
plt.scatter(important_cols["first_target_pct"], important_cols["result"])
```

```python
important_cols["ambition"] = 0.0
important_cols["realization"] = 0.0

for i in important_cols.index:
   
    state = important_cols.loc[i, "Unnamed: 0.1"]
    year = important_cols.loc[i, "nearest_year"]
    print(state)
    
    target_year = important_cols.loc[i, "first_target_year"]
    start_year = important_cols.loc[i, "year_enacted"]

    progress = (year - start_year) / (target_year - start_year)

    # initial conditions calculation 
    state_array = re_prop(state)
    start_row = state_array[state_array.year == start_year]
    curr_row = state_array[state_array.year == year]

    start_ratio = start_row["RETCB"]/start_row["TETCB"]
    target_ratio = important_cols.loc[i, "first_target_pct"] 
    difference = target_ratio - start_ratio

    year_difference = target_year - start_year
    curr_difference = year - start_year

    ambition = difference / year_difference

    ratio = curr_row["RETCB"]/curr_row["TETCB"]

    curr_ratio_difference = (float(ratio) - float(start_ratio))
    realization = curr_ratio_difference / curr_difference

    print("Target: ", target_ratio)
    print("Start: ", start_ratio)
    print("Difference between target and start: ", difference)
    print("ambition:", year_difference, ambition)
    print("current ratio: ", ratio)
    print("Difference between start and now: ", curr_difference)
    print("realization", curr_difference, realization)

    important_cols.at[i, "ambition"] = ambition
    important_cols.at[i, "realization"] = realization
    important_cols

    
```
